```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)

# Fill missing Embarked with mode
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# Drop Cabin column (too many missing)
df.drop('Cabin', axis=1, inplace=True)
```

```python
sns.set(style="whitegrid")

# 1. Survival Count
sns.countplot(x='Survived', data=df)
plt.title('Survival Count')
plt.show()
```

```python
# 2. Survival by Gender
sns.countplot(x='Survived', hue='Sex', data=df, palette='Set2')
plt.title('Survival by Gender')
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Survived', data=df, palette='pastel')
plt.title('Passenger Class vs Survival')
plt.show()
```

```python
sns.histplot(data=df, x='Age', hue='Survived', bins=30, kde=True, palette='Dark2')
plt.title('Age Distribution by Survival')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.select_dtypes(include=np.number).corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.countplot(x='Embarked', hue='Survived', data=df, palette='viridis')
plt.title('Survival by Embarkation Port')
plt.xlabel('Embarked (C = Cherbourg, Q = Queenstown, S = Southampton)')
plt.ylabel('Count')
plt.show()
```
